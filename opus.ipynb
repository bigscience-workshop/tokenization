```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url="https://opus.nlpl.eu/CCAligned-v1.php"

# Make a GET request to fetch the raw HTML content
html_content = requests.get(url).text

# Parse the html content
soup = BeautifulSoup(html_content, "lxml")
# print(soup.prettify()) # print the parsed data of html
```

```python
# https://opus.nlpl.eu/CCAligned/v1/ar-en_sample.html
import re
from collections import defaultdict

count = 0
html_content = []

en = 'en'
columns = [str(i.get("href"))[13:15] for i in soup.find_all("a") if '-en_sample.html' in str(i.get("href")) ] + [ str(i.get("href"))[16:18] for i in soup.find_all("a") if 'en-' in str(i.get("href")) and 'CN' or 'TW' not in str(i.get("href")) ] + [ str(i.get("href"))[16:21] for i in soup.find_all("a") if 'en-' in str(i.get("href")) and 'CN' or 'TW' in str(i.get("href")) ]
columns.insert(0,en)

data = {k: [] for k in columns}
eng_other = {k: [] for k in columns}
del eng_other["CA"]
del eng_other["CAlig"]
print(eng_other)
```

```python

for link in soup.find_all("a"):
  if '-en_sample.html' in str(link.get("href")):
    print(link.get("href"))
    href = "https://opus.nlpl.eu/" + str(link.get("href"))
    html = str(BeautifulSoup(requests.get(href).text, "lxml"))
    html_split = html.split("<hr/>")
    html_split_regex = []
    html_split = html_split[1:]
    for h in html_split:
      splits = h.split("\n")
      src = splits[1]
      targ = splits[2]
      if "<br/>" in h:
        src = re.search(';(.*)<', src).group(0)[1:-1]
        targ = re.search(';(.*)<', targ).group(0)[1:-1]
        # if targ not in eng_other['en']:
        eng_other['en'].append(targ)
        eng_other[str(link.get("href"))[13:15]].append(src)
        for k in list(eng_other.keys()):
          if k != "en" and k != str(link.get("href"))[13:15]:
            eng_other[k].append(None)

        


```
